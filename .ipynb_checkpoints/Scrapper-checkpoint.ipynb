```python
import requests

url = 'https://www.deped.gov.ph/contact-us/regional-division-offices-directory/'
headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)
if response.status_code == 200:
    html_content = response.text
else:
    print(f'Failed to retrieve the page. Status code: {response.status_code}')
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html_content, 'html.parser')
```

```python

names_emails = []

for row in soup.select('table tbody tr'):
    division = row.select('td')[0].get_text(strip=True)
    superintendent = row.select('td')[1].get_text(strip=True)
    assistant_superintendent = row.select('td')[2].get_text(strip=True)
    contact_info = row.select('td')[3].get_text(strip=True)
    email = None
    if '@' in contact_info:
        email = contact_info.split()[-1]
    if email:
        names_emails.append({
            'division': division,
            'superintendent': superintendent,
            'assistant_superintendent': assistant_superintendent,
            'email': email
        })

for item in names_emails:
    print(f"Divsion:{item['division']} \n Superintendent: {item['superintendent']}, \n Assistant: {item['assistant_superintendent']}, \n Email: {item['email']}")
```

```python
import csv

filename = "div_superintendents_and_emails.csv"

fieldnames = ['division','superintendent', 'assistant_superintendent', 'email']

# Writing to CSV
with open(filename, mode='w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    
    # Write the header (column names)
    writer.writeheader()

    for item in names_emails:
        writer.writerow(item)

print(f"Data saved to {filename}")
```
